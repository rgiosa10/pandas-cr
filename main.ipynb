```python
# Run shape again to see number of rows before removing duplicate rows:
print(albums_df.shape) 
print(artists_df.shape)
print(tracks_df.shape)

#Clean and normalize each dataset by remove duplicates from each:
albums_df.drop_duplicates(inplace=True)
artists_df.drop_duplicates(inplace=True)
tracks_df.drop_duplicates(inplace=True)

# Run shape again to see number of rows after removing duplicate rows. 
# Data shows that there were no duplicate rows to remove as number of rows remains the same
print(albums_df.shape) 
print(artists_df.shape)
print(tracks_df.shape)
```

```python
# Replace the empty list '[]' entries in the genres column with NaNs

def replace_with_NAN(value):
    if value == '[]':
        value = np.nan
        return value
    else:
        return value

artists_df.genres = artists_df.genres.map(replace_with_NAN)

#provides sample of genres column data to validate that replaced '[]' with NaN
print(artists_df.genres.sample(10))  
```

```python
# join artists and albums on artist ID:
albums_df = pd.read_csv('./data/spotify_spotify_albums.csv', header=0,)
tracks_df = pd.read_csv('./data/spotify_spotify_tracks.csv', header=0)

# Data profiling:
print(albums_df.shape) 
print(tracks_df.shape)

# Data Cleaning:
albums_df['id'].dropna(inplace=True) 
tracks_df['album_id'].dropna(inplace=True)

albums_df['id']= albums_df['id'].astype(str) 
tracks_df['album_id']= tracks_df['album_id'].astype(str)

print(albums_df['artist_id'].dtypes)
print(artists_df['id'].dtypes)

# using outer merge to consolidate and then checking to see if there are any duplicates ids
album_tracks_right_merge = pd.merge(albums_df, tracks_df, how='right', left_on='id', right_on='album_id')

# Used a merge of album data on to the tracks as I expect multiple tracks could fall on one album.
# So it would be better to pair the albums data to tracks data.

print(album_tracks_right_merge.head())
print(album_tracks_right_merge.shape)

# Below commented out code is not working for the join:
#album_tracks_left_join = tracks_df.join(albums_df.set_index('id'), how='left', on='album_id')
# Could not figure out why it would not match up. I believe it would be an issue with data type but could not figure out way to get 'id' and 'album_id' to convert to other data type than object.
```

```python
# Which artists appear the most times in Artists data
print(artists_df.groupby(by='name').name.count().sort_values(ascending=False))

# Answer: based on above print it shows that the artists: Haze and Sasha show up each 5 times
```

```python
# Which artists have the highest 'artist_popularity' rankings. List top ten in descending order:
print(artists_df.groupby(by='name').artist_popularity.max().sort_values(ascending=False).head(10))
```

```python
# Bonus Point section:
# Number of albums that came out in each year

# Creates a new column with just the year (not full date) that album released
albums_df['year_release_date'] = pd.DatetimeIndex(albums_df['release_date']).year

# Groups by year released and counts how many
print(albums_df.groupby(by='year_release_date').year_release_date.count())
```
